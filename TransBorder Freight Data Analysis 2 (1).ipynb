```python
import pandas as pd
import os

# to suppress warnings
import warnings
warnings.filterwarnings('ignore')

# Define the base directory and months
base_dir = "C:\\Users\\LENOVO\\Desktop\\Azubi Africa TMP\\data\\2023"
months = ["Jan2023", "Feb2023", "March2023", "April2023", "May2023", "June2023", "July2023", "Aug2023", "sept2023", "Oct2023", "Nov2023", "December2023"]

# Dictionary to store combined data for each month
monthly_data_2023 = {}

for month in months:
    # Create the path for the month's folder
    month_dir = os.path.join(base_dir, month, month)
    
    # Load and concatenate all CSV files in the month's folder
    files = [f for f in os.listdir(month_dir) if f.endswith(".csv")]
    dataframes = [pd.read_csv(os.path.join(month_dir, file)) for file in files]
    monthly_data_2023[month] = pd.concat(dataframes, ignore_index=True)

print(monthly_data_2023)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# A scatter plot for the relationship  between shipment weight and freight charges
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='SHIPWT', y='FREIGHT_CHARGES', hue='TRDTYPE', alpha=0.6)
plt.title('Relationship Between Shipment Weight and Freight Charges')
plt.xlabel('Shipment Weight')
plt.ylabel('Freight Charges')
plt.legend(title='Transportation Mode')
plt.tight_layout()
plt.show()
```

```python
# A bar plot for average freight charges by state and trade type 
plt.figure(figsize=(14, 7))
sns.barplot(data=data, x='USASTATE', y='FREIGHT_CHARGES', hue='TRDTYPE', estimator='mean', ci=None)
plt.title('Average Freight Charges by State and Trade Type')
plt.xlabel('State')
plt.ylabel('Average Freight Charges')
plt.xticks(rotation=45)
plt.legend(title='Trade Type')
plt.tight_layout()
plt.show()
```

```python
# A line plot for yearly trends in freight movement by trade type
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='YEAR', y='VALUE', hue='TRDTYPE', marker='o')
plt.title('Yearly Trends in Freight Movement by Trade Type')
plt.xlabel('Year')
plt.ylabel('Total Freight Value')
plt.xticks(rotation=45)
plt.legend(title='Trade Type')
plt.tight_layout()
plt.show()
```

```python
# A line plot for yearly trends in freight movement by transportation mode 
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='YEAR', y='VALUE', hue='DISAGMOT', marker='o')
plt.title('Yearly Trends in Freight Movement by Transportation Mode')
plt.xlabel('Year')
plt.ylabel('Total Freight Value')
plt.xticks(rotation=45)
plt.legend(title='Transportation Mode')
plt.tight_layout()
plt.show()
```

```python
# A heatmap to visualize underutilised transportation routes
utilization_data = data.pivot_table(index='DISAGMOT', columns='TRDTYPE', values='VALUE', aggfunc='sum')  
plt.figure(figsize=(12, 8))
sns.heatmap(utilization_data, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Utilization of Transportation Routes')
plt.xlabel('Trade Type')
plt.ylabel('Route')
plt.show()
```

```python
# A bar plot to calculate average freight charges by mode of transportation 
plt.figure(figsize=(10, 5))
sns.barplot(data=data, x='DISAGMOT', y='FREIGHT_CHARGES')
plt.title('Average Freight Charges by Mode of Transportation')
plt.ylabel('Average Freight Charges')
plt.xticks(rotation=45)
plt.show()
```
